## 3. ***Data Wrangling***

First- We will Create some new column in our df. These column will help us to indentify charges and plan issues
New columns will be:-
1. Call Duration - to check the average duration per call of each customer
2. Call charge - to check the average charge per minute
3. Retained Category- To categorised the data in the month of custome retained
4. Area Name - To check the Area Name of the Area Code

By creating these column We will not require the Area code, Total minutes, Total call and Total Charges

````python
#creating copy of Dataset, we will not edit our original data
df=dataset.copy()
````

####Creating some columns


````python
#Creating Day call duration column
df['Day Call Duration']=dataset['Total day minutes']/dataset['Total day calls']
#Creating Day call Charge column
df['Day Call Charge']=dataset['Total day charge']/dataset['Total day minutes']

#Creating Evening call duration column
df['Evening call duration']=dataset['Total eve minutes']/dataset['Total eve calls']
#Creating Evening call Charge column
df['Evening call Charge']=dataset['Total eve charge']/dataset['Total eve minutes']

#Creating Night Call Duration column
df['Night Call Duration']=dataset['Total night minutes']/dataset['Total night calls']
#Creating Night Call Charge column
df['Night Call Charge']=dataset['Total night charge']/dataset['Total night minutes']

````

````python
#categorised account length column
Account_length_category=[]
for i in df['Account length']:
  if i>0 and i<31:
    Account_length_category.append("Between 1 to 30 Days")
  elif i>30 and i<91:
    Account_length_category.append("Between 31 to 90 Days")
  elif i>90 and i<121:
    Account_length_category.append("Between 91 to 120 Days")
  elif i>120 and i<181:
    Account_length_category.append("Between 121 to 180 Days")
  elif i>180 and i<366:
    Account_length_category.append("More than 6 month")
  else:
    Account_length_category.append("More than 1 year")

Area_list=[]
#Rename the code by area Name
for i in df['Area code']:
  if i==415:
    Area_list.append("San Francisco")
  elif i==408:
    Area_list.append("San Jose")
  elif i==510:
    Area_list.append("City of Oakland")
  else:
    Area_list.append("Not Available")


#creating two column Area Name and Retained Category
df['Retained Category']=Account_length_category
df['Area Name']=Area_list
df['New State Name']=df['State']+"-"+df['Area Name']
````

####Deleting some new columns

````python
#Deleting Extra columns
columns_to_delete=['Area code', 'Total day minutes', 'Total day calls', 'Total day charge', 'Total eve minutes', 'Total eve calls', 'Total eve charge', 'Total night minutes', 'Total night calls', 'Total night charge']
for i in columns_to_delete:
  if i in df.columns:
    del df[i]
  else:
    print("Columns not available")

#adding new column of total average charges
df['Total avg Charges']=df['Day Call Charge']+df['Evening call Charge']+df['Night Call Charge']

#adding new column of total average charges
df['Total Average Duration']=df['Day Call Duration']+df['Evening call duration']+df['Night Call Duration']
````
